```python
def number_unique_values(data):
    """Count the number of unique values for each column
    
    Return a DataFrame with the name of the columns, the type and the number of unique values
    """
    res = pd.DataFrame([(name, data[name].dtype.name, data[name].nunique()) for name in data],
                      columns=['name', 'type', 'number'])
    return res
```

```python
y=Dataset['becameLicensed']
```

```python
Finala=Dataset.drop(['becameLicensed', 'Campaign', 'Segment', 'teamTrialOwner'], axis=1)
```

```python
Finala=Finala.drop(['Country', 'Language', 'wasSharedToPriorToTrial'], axis=1)
```

```python
Finala=Finala.drop(['mobileLoginsCount', 'attachmentCount', 'LoginsCount',  'insertColumnsCount', 'attachmentCount',  'reportCount', 'discussionCount'], axis=1)
```

```python
Finala.head()
```

```python
from sklearn import preprocessing

x1 = Finala.values #returns a numpy array
min_max_scaler = preprocessing.MinMaxScaler()
x1_scaled = min_max_scaler.fit_transform(x1)
Finala1 = pd.DataFrame(x1_scaled)
```

```python
Finala1.head()
```

```python
X=Finala1
```

```python
from sklearn.cross_validation import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, random_state=1)
print (X_train.shape)
print (X_test.shape)
```

```python
from sklearn.linear_model import LogisticRegression
LR=LogisticRegression()
LR.fit(X_train,y_train)
```

```python
y_pred_LR=LR.predict(X_test)
y_pred_proba_LR=LR.predict_proba(X_test)
```

```python
from sklearn import metrics
print (metrics.accuracy_score(y_test, y_pred_LR)) 
```

```python
metrics.confusion_matrix(y_test, y_pred_LR)
```
